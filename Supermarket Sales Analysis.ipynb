```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_excel(r'C:\Users\Ulum\Downloads\Supermarket-Sales-Sample-Data.xlsx')
df.head(20)
```

```python
header = df.iloc[6]
```

```python
df=df[7:]
df.columns = header
df.head()
```

```python
df.reset_index(inplace=True)
df.head()
```

```python
df.drop(columns=df.columns[0], inplace=True)
df.head()
```

```python
df.info()
```

```python
df['Order No'] = df['Order No'].astype(int)
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d/%m/%Y')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format = '%d/%m/%Y')
df['Retail Price (USD)'] = df['Retail Price (USD)'].astype(float) 
df['Order Quantity'] = df['Order Quantity'].astype(int)
df['Tax (USD)'] = df['Tax (USD)'].astype(float)
df['Total (USD)'] = df['Total (USD)'].astype(float)
df.info()
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df.duplicated().sum()
```

```python
df['Processed Time'] = (df['Ship Date'] - df['Order Date']).dt.days
```

```python
df.head()
```

```python
df['Customer Total Sales'] = df.groupby('Customer Name')['Total (USD)'].transform('sum')
df['Customer Order Count'] = df.groupby('Customer Name')['Order No'].transform('nunique')
df['Average Customer Order'] = round(df.groupby('Customer Name')['Total (USD)'].transform('mean'),3)
```

```python
round(df.describe(),2)
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Customer Total Sales'], bins=30,kde=True)
plt.title('Total Sales Distribution')
plt.xlabel('Total Sales')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Processed Time'], bins=10,kde=True)
plt.title('Processed Time Distribution')
plt.xlabel('Processed Time')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=df['Processed Time'],y=df['Total (USD)'], alpha=0.5)
plt.ylabel('Total Sales (USD)')
plt.xlabel('Processed Time')
    
plt.plot()
```
